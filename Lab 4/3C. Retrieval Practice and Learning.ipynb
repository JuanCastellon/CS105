# Retrieval Practice and Learning

What is the most effective way to learn a subject? Many students focus exclusively on the _encoding_ process---that is, how to get the knowledge into memory in the first place. For example, taking notes is an activity for encoding knowledge.

_Retrieval_, on the other hand, is the process of reconstructing that knowledge from memory. [Karpicke and Blunt](http://science.sciencemag.org/content/331/6018/772) (2011) demonstrated that _retrieval_ is more effective for learning than activites designed to promote effective encoding. They conducted an experiment in which subjects had to learn about sea otters by reading a passage. Subjects were randomly assigned to one of two conditions: some were instructed to create a [concept map](https://en.wikipedia.org/wiki/Concept_map) as they read the passage, while others were instructed to practice retrieval (i.e., read the passage, recall as much as they could, read the text again, and recall again). The two main measurements they recorded were:

1. each subject's score on a follow-up learning test one week later
2. each subject's _prediction_ of how well they would do on that test

In this lab, you will analyze data from a _replication_ of Karpicke and Blunt's experiment, conducted by Buttrick _et al_.
- The data file is : data.csv.
- The codebook (explaining what the variables mean) is : codebook.csv.

# Question 3

How good were subjects at predicting how well they would do on the follow-up learning test? Calculate a measure of how well subjects predicted their performance and interpret the value in context. (Optionally, you may want to include a visualization as well.)

```python
# YOUR CODE HERE
concept_correct = df[["Condition","PR.2","TS.avg"]].where(df["Condition"] == "Concept")
concept_correct = concept_correct.dropna()
# This is the calculation is for detecting the discrepancy between a subject's 
conceptTotal = (concept_correct["TS.avg"].mean() - concept_correct["PR.2"].mean()) # / len(concept_correct.index)

retrieval_correct = df[["Condition","PR.2","TS.avg"]].where(df["Condition"] == "Retrieval")
retrieval_correct = retrieval_correct.dropna()
retrievalTotal = (retrieval_correct["TS.avg"].mean() - retrieval_correct["PR.2"].mean()) # / len(retrieval_correct.index)

prediction = df.groupby("Condition")[["PR.2","TS.avg"]].mean().plot.bar()
print("Concept Comparison = ",conceptTotal)
print("Retrieval Comparison = ",retrievalTotal)
```

**YOUR EXPLANATION HERE**

The calculation we're making here is comparing a subject's predicted score (PR.2) and their actual average test score (TS.avg) of their follow up learning test. The calculation itself is getting the mean of the subjects' actual test averages and subtracting their predicted score from them. 
If the predicted scores of a subject matches their actual scores, they will have a 'correctness' score of 0.
Likewise, if the subject performed worse than their predicted score, their score will be less than 0.
With this calculation we can see that the retrieval correctness score was much closer to 0 than the concept correctness score. This also shows that retrieval learners performed better than their predictions. The graph below also shows that the concept group overestimated how well they'd do on the follow up test compared to the retrieval group.

# Question 4

This was a completely randomized experiment. This means that the condition that each subject was assigned to should be independent of their gender, age, and any other subject characteristics. Does that seem to be true in this case? Calculate a summary measure and/or make a visualization, and explain what you see.

```python
# YOUR CODE HERE
independent_1 = pd.crosstab(df["Gender"], df["Condition"])
independent_2 = pd.crosstab(df["Age"], df["Condition"])
print(independent_1)
print(independent_2)
independent_1.plot.bar()
independent_2.plot.bar()
```

**YOUR EXPLANATION HERE**

Males are overwhelmingly more likely to be in the concept group instead of the retrieval group. Females are a little more balanced in that regard. 18 years olds are overwhelmingly more likely to be apart of the concept group as well. In general, both males and females lean more towards being conceptual than retrieval. Likewise, being younger (18-20) gives you a much higher chance of being in the concept group than the retrieval group. It would seem that gender and age are not totally independent of the category they are assigned.
